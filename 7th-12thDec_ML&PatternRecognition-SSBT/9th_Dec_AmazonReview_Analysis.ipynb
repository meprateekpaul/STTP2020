```
import nltk
nltk.download("vader_lexicon")
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```
import pandas as pd
```

```
data = pd.read_csv("/content/amazon_reviews.tsv",sep="\t")
data.head(20)
```

```
data.iloc[0]['review']
```

```
data.iloc[0,1]
```

```
sia.polarity_scores(data.iloc[0]['review'])
```

```
len(data)
```

```
polarity=[]
for i in range(len(data)):
  polarity.append(sia.polarity_scores(data.iloc[i]['review']))
```

```
polarity
```

```
data['Polarity Scores'] = polarity
data.head()
```

```
data.iloc[0]["Polarity Scores"]['compound']
```

```
compound =[]
for i in range(len(data)):
  compound.append(data.iloc[i]["Polarity Scores"]['compound'])
```

```
data['Compound'] = compound
```

```
data.head()
```

```
data['label'].value_counts()
```

```
sentiment = []
for i in range(len(data)):
  c = data.iloc[i]['Compound']
  if c>0:
    sentiment.append("pos")
  else:
    sentiment.append("neg")
```

```
data['Predicted Label'] = sentiment
data.head(10)
```

```
### Comparision
```

```
count = 0
for i in range(len(data)):
  label = data.iloc[i]['label']
  pred_label = data.iloc[i]['Predicted Label']
  if label == pred_label:
    count=count+1
```

```
count
```

```
accuracy = count/len(data)*100
accuracy
```

```
###ConfusiomMatrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(data['label'],data['Predicted Label'])
cm
```

```
###Accuracy score
from sklearn.metrics import accuracy_score
a = accuracy_score(data['label'],data['Predicted Label'])
a*100
```
