```
###Face Detection
img = cv2.imread("/content/face.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
face_cascade = cv2.CascadeClassifier("/content/face.xml")
faces = face_cascade.detectMultiScale(img) #detecting anything based on the above xml file
for x,y,w,h in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,255),5)
plt.imshow(img)
plt.show()
```

```
### eye Detection
img = cv2.imread("/content/face.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
face_cascade = cv2.CascadeClassifier("/content/face.xml")
eye_cascade = cv2.CascadeClassifier("/content/eye.xml")
faces = face_cascade.detectMultiScale(img)
for x,y,w,h in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),5)
  eyes = eye_cascade.detectMultiScale(img[y:y+h,x:x+w],1.22) #() reduce the pixel values, reducing the error
  for ex,ey,ew,eh in eyes:
    cv2.rectangle(img[y:y+h,x:x+w],(ex,ey),(ex+ew,ey+eh),(255,0,0),3)
plt.imshow(img)
plt.show()
```
